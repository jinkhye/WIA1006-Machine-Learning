Analyse target variables

```python
# Plot the distribution
sns.histplot(data['price'] , kde=True);
plt.axvline(x = data['price'].mean(), c = 'red')
plt.axvline(x = data['price'].median(), c = 'green')
plt.ylabel('Frequency')
plt.title('Price distribution')

# Plot the QQ-plot
fig = plt.figure()
res = stats.probplot(data['price'], plot=plt)
plt.show()
```

The target variable is right skewed so we will do log transformation to better fit the normal distribution

```python
# Log transformation of target variable
data["price"] = np.log1p(data["price"])

# Check the new distribution
sns.histplot(data['price'] , kde=True);
plt.axvline(x = data['price'].mean(), c = 'red')
plt.axvline(x = data['price'].median(), c = 'green')
plt.ylabel('Frequency')
plt.title('Price distribution')

# Plot the QQ-plot
fig = plt.figure()
res = stats.probplot(data['price'], plot=plt)
plt.show()
```

Outlier Analysis

```python
# Plot the boxplots
fig, axs = plt.subplots(2,3, figsize = (10,5))
plt1 = sns.boxplot(data['price'], ax = axs[0,0])
plt2 = sns.boxplot(data['area'], ax = axs[0,1])
plt3 = sns.boxplot(data['bedrooms'], ax = axs[0,2])
plt1 = sns.boxplot(data['bathrooms'], ax = axs[1,0])
plt2 = sns.boxplot(data['stories'], ax = axs[1,1])
plt3 = sns.boxplot(data['parking'], ax = axs[1,2])

plt.tight_layout()
```

There is outlier in area.

```python
# Outlier treatment for area
Q1 = data.area.quantile(0.25)
Q3 = data.area.quantile(0.75)
IQR = Q3 - Q1
data = data[(data.area >= Q1 - 1.5*IQR) & (data.area <= Q3 + 1.5*IQR)]
plt.boxplot(data.area)
data = data.reset_index(drop=True)
```

```python
# Check the type of data
data.dtypes
```

```python
# Identify the categorical and numerical features
cat_data = ["mainroad","guestroom","basement","hotwaterheating","airconditioning","prefarea","furnishingstatus"]
num_data = ["area","bedrooms","bathrooms","stories","parking"]
```

```python
# Label encoding to convert categorical features into numerical values
lb = LabelEncoder()
for i in cat_data:
    data[i] = lb.fit_transform(data[i])

data.head()
```

```python
# Rescale the features
scaler = MinMaxScaler()
data[num_data] = scaler.fit_transform(data[num_data])
data.head()
```

```python
# Correlation heatmap
plt.figure(figsize=(7,7))
sns.heatmap(data.corr(),annot=True);
```

```python
# Separate the features and target variable
X = data.drop("price", axis=1)
y = data["price"]
```

```python
# Split the data into training and testing sets
X_train, X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state = 100)
```

## Regression Model Development

```python
# Rescaling
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
print("X_train",X_train.shape)
print("y_train",y_train.shape)
print("X_test",X_test.shape)
print("y_test",y_test.shape)
```

```python
# Define functions for evaluation

def print_evaluate(true, predicted):  
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2score = r2_score(true, predicted)
    print('MAE:', mae)
    print('MSE:', mse)
    print('RMSE:', rmse)
    print('R2 Score:', r2score)
    
def evaluate(true, predicted):
    mae = mean_absolute_error(true, predicted)
    mse = mean_squared_error(true, predicted)
    rmse = np.sqrt(mean_squared_error(true, predicted))
    r2score = r2_score(true, predicted)
    return mae, mse, rmse, r2score
```

Linear Regression without Regularization

```python
# Build linear regression model
lr = LinearRegression()

# Train the model
lr.fit(X_train,y_train)

# Training accuracy of model
print("Training score:",lr.score(X_train,y_train),end='\n')

# Test accuracy of model
print("Testing score:",lr.score(X_test,y_test))

# Evaluate training and testing set
ypred_train = lr.predict(X_train)
ypred_test = lr.predict(X_test)

print('\nTrain set evaluation:\n')
print_evaluate(y_train,ypred_train)
print('\nTest set evaluation:\n')
print_evaluate(y_test, ypred_test)
```

```python
# Plot predictions
y_train = np.expm1(y_train)
ypred_train = np.expm1(ypred_train)
y_test = np.expm1(y_test)
ypred_test = np.expm1(ypred_test)

fig = plt.figure()
plt.scatter(y_test, ypred_test)
plt.title('Predicted vs Actual House Prices (Linear Regression)')            
plt.xlabel('Actual')                         
plt.ylabel('Predicted')  
plt.plot([1000000, 13000000], [1000000, 13000000], c = "red")
```

```python
# Predictions
y_test.shape
y_test_matrix = y_test.values.reshape(-1,1)

per_error = 100*(ypred_test-y_test)/y_test

# Load actual and predecited values side by side
dframe=pd.DataFrame({'actual':y_test_matrix.flatten(),'Predicted':ypred_test.flatten(), '% Error':per_error}) 

# Flatten to get single axis of data (1 dimension only)
dframe.head(15)
```

```python
# Plot actual and predicted values
plt.figure(figsize=(18,6))
plt.plot((ypred_test)[:100]) 
plt.plot((np.array(y_test)[:100]))
plt.legend(["Predicted","Actual"])
plt.title('Actual and Predicted Time Duration')
```

2* Linear Regression with Ridge regularization

```python
y_train = np.log1p(y_train)
ypred_train = np.log1p(ypred_train)
y_test = np.log1p(y_test)
ypred_test = np.log1p(ypred_test)

# Ridge regularization with different values of alpha
ridge = RidgeCV(alphas = [0.01, 0.03, 0.06, 0.1, 0.3, 0.6, 1, 3, 6, 10, 30, 60])
ridge.fit(X_train, y_train)

# Best alpha parameter
alpha = ridge.alpha_
print("Best alpha :", alpha)

print("Try again for more precision with alphas centered around " + str(alpha))
ridge = RidgeCV(alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, alpha * .85, 
                          alpha * .9, alpha * .95, alpha, alpha * 1.05, alpha * 1.1, alpha * 1.15,
                          alpha * 1.25, alpha * 1.3, alpha * 1.35, alpha * 1.4], 
                cv = 10)
ridge.fit(X_train, y_train)
alpha = ridge.alpha_
print("Best alpha :", alpha)

y_train_rdg = ridge.predict(X_train)
y_test_rdg = ridge.predict(X_test)

# Training accuracy of model
print("\nThe training score is",ridge.score(X_train,y_train_rdg))

# Test accuracy of model
print("The testing score is",ridge.score(X_test,y_test_rdg))

print('\nTrain set evaluation:\n')
print_evaluate(y_train,y_train_rdg)
print('\nTest set evaluation:\n')
print_evaluate(y_test, y_test_rdg)

y_train = np.expm1(y_train)
y_train_rdg = np.expm1(ypred_train)
y_test = np.expm1(y_test)
y_test_rdg = np.expm1(ypred_test)

# Plot predictions
fig = plt.figure()
plt.scatter(y_test, y_test_rdg)
plt.title('Linear Regression with Ridge Regularization')          
plt.xlabel('Actual')                         
plt.ylabel('Predicted')  
plt.plot([1000000, 13000000], [1000000, 13000000], c = "red")
plt.show()
```

3* Linear Regression with Lasso regularization

```python
# 3* Lasso
y_train = np.log1p(y_train)
y_test = np.log1p(y_test)

lasso = LassoCV(alphas = [0.0001, 0.0003, 0.0006, 0.001, 0.003, 0.006, 0.01, 0.03, 0.06, 0.1, 
                          0.3, 0.6, 1], 
                max_iter = 50000, cv = 10)
lasso.fit(X_train, y_train)

# best alpha parameter
alpha = lasso.alpha_
print("Best alpha :", alpha)

print("Try again for more precision with alphas centered around " + str(alpha))
lasso = LassoCV(alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, 
                          alpha * .85, alpha * .9, alpha * .95, alpha, alpha * 1.05, 
                          alpha * 1.1, alpha * 1.15, alpha * 1.25, alpha * 1.3, alpha * 1.35, 
                          alpha * 1.4], 
                max_iter = 50000, cv = 10)
lasso.fit(X_train, y_train)

alpha = lasso.alpha_
print("Best alpha :", alpha)

y_train_las = lasso.predict(X_train)
y_test_las = lasso.predict(X_test)

# Training accuracy of model
print("\nThe training score is",lasso.score(X_train,y_train_las))

# Test accuracy of model
print("The testing score is",lasso.score(X_test,y_test_las))

# Evaluate the training and test set
print('\nTrain set evaluation:\n')
print_evaluate(y_train,y_train_las)
print('\nTest set evaluation:\n')
print_evaluate(y_test, y_test_las)


y_train = np.expm1(y_train)
y_train_las = np.expm1(ypred_train)
y_test = np.expm1(y_test)
y_test_las = np.expm1(ypred_test)

# Plot predictions
fig = plt.figure()
plt.scatter(y_test, y_test_las)
plt.title('Linear Regression with Lasso Regularization')          
plt.xlabel('Actual')                         
plt.ylabel('Predicted')  
plt.plot([1000000, 13000000], [1000000, 13000000], c = "red")
plt.show()
```

4* Linear Regression with ElasticNet regularization

```python
y_train = np.log1p(y_train)
y_test = np.log1p(y_test)

elasticNet = ElasticNetCV(l1_ratio = [0.1, 0.3, 0.5, 0.6, 0.7, 0.8, 0.85, 0.9, 0.95, 1],
                          alphas = [0.0001, 0.0003, 0.0006, 0.001, 0.003, 0.006, 
                                    0.01, 0.03, 0.06, 0.1, 0.3, 0.6, 1, 3, 6], 
                          max_iter = 50000, cv = 10)
elasticNet.fit(X_train, y_train)
alpha = elasticNet.alpha_
ratio = elasticNet.l1_ratio_
print("Best l1_ratio :", ratio)
print("Best alpha :", alpha )

print("Try again for more precision with l1_ratio centered around " + str(ratio))
elasticNet = ElasticNetCV(l1_ratio = [ratio * .85, ratio * .9, ratio * .95, ratio, ratio * 1.05, ratio * 1.1, ratio * 1.15],
                          alphas = [0.0001, 0.0003, 0.0006, 0.001, 0.003, 0.006, 0.01, 0.03, 0.06, 0.1, 0.3, 0.6, 1, 3, 6], 
                          max_iter = 50000, cv = 10)
elasticNet.fit(X_train, y_train)
if (elasticNet.l1_ratio_ > 1):
    elasticNet.l1_ratio_ = 1    
alpha = elasticNet.alpha_
ratio = elasticNet.l1_ratio_
print("Best l1_ratio :", ratio)
print("Best alpha :", alpha )

print("Now try again for more precision on alpha, with l1_ratio fixed at " + str(ratio) + 
      " and alpha centered around " + str(alpha))
elasticNet = ElasticNetCV(l1_ratio = ratio,
                          alphas = [alpha * .6, alpha * .65, alpha * .7, alpha * .75, alpha * .8, alpha * .85, alpha * .9, 
                                    alpha * .95, alpha, alpha * 1.05, alpha * 1.1, alpha * 1.15, alpha * 1.25, alpha * 1.3, 
                                    alpha * 1.35, alpha * 1.4], 
                          max_iter = 50000, cv = 10)
elasticNet.fit(X_train, y_train)
if (elasticNet.l1_ratio_ > 1):
    elasticNet.l1_ratio_ = 1    
alpha = elasticNet.alpha_
ratio = elasticNet.l1_ratio_
print("Best l1_ratio :", ratio)
print("Best alpha :", alpha )

y_train_ela = elasticNet.predict(X_train)
y_test_ela = elasticNet.predict(X_test)

# Training accuracy of model
print("\nThe training score is",elasticNet.score(X_train,y_train_ela))

# Test accuracy of model
print("The testing score is",elasticNet.score(X_test,y_test_ela))

# Evaluate the training and test set
print('\nTrain set evaluation:\n')
print_evaluate(y_train,y_train_ela)
print('\nTest set evaluation:\n')
print_evaluate(y_test, y_test_ela) 

y_train = np.expm1(y_train)
y_train_ela = np.expm1(ypred_train)
y_test = np.expm1(y_test)
y_test_ela = np.expm1(ypred_test)

# Plot predictions
fig = plt.figure()
plt.scatter(y_test, y_test_ela)
plt.title('Linear Regression with ElasticNet Regularization')          
plt.xlabel('Actual')                         
plt.ylabel('Predicted')  
plt.plot([1000000, 13000000], [1000000, 13000000], c = "red")
plt.show()
```

By comparing all the models, ridge regression model has slightly lower RMSE and higher R2 score compared to the linear regression without regularization. Lasso and elastic net regularization do not give signicant improvement to the model. We can conclude that the ridge regression model gives the best result.
